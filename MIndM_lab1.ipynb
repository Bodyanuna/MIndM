```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# 1. Створення даних
n = 18
np.random.seed(42)

x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 0.5, size=x.shape)
y = n * x + np.sin(x / n) + noise

# 2. Розділення даних
X = x.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Побудова моделі
model = LinearRegression()
model.fit(X_train, y_train)

a = model.coef_[0]
b = model.intercept_
print(f"Коефіцієнти: a = {a:.4f}, b = {b:.4f}")

# 4. Оцінка якості
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE = {mse:.4f}")
print(f"MAE = {mae:.4f}")
print(f"R² = {r2:.4f}")

# 5. Графік
plt.scatter(X, y, color="blue", label="Дані")
plt.plot(X, model.predict(X), color="red", label="Лінійна регресія")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Лінійна регресія")
plt.legend()
plt.show()

# 6. Висновок
if r2 > 0.9:
    print("Модель дуже добре описує дані.")
elif r2 > 0.7:
    print("Модель досить добре описує дані, але є нелінійності.")
else:
    print("Модель погано підходить для цих даних, оскільки залежність не зовсім лінійна.")
```
